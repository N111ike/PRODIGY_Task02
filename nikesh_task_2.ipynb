```python
import pandas as pd
```

```python
file_path = '/content/titanic-publicleaderboard-2025-04-13T10_13_31.csv'
```

```python
df = pd.read_csv(file_path)
```

```python
df.isnull().sum()
```

```python
# Fill Age with median
df['TeamId'].fillna(df['TeamId'].median(), inplace=True)
```

```python
df['LastSubmissionDate']=pd.to_datetime(df['LastSubmissionDate'])
```

```python
mode_date = df['LastSubmissionDate'].mode()[0]
```

```python
df['LastSubmissionDate'].fillna(mode_date, inplace=True)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.countplot(data=df, x='TeamName', hue='Score')
```

```python
plt.show()
```

```python
plt.title("Teamname and score")
```

```python
print(df.columns)
```

```python
sns.countplot(data=df, x='Rank', hue='Score')
```

```python
sns.histplot(data=df, x='Score', bins=30, kde=True)
```

```python
import numpy as np
```

```python
df_gender = pd.DataFrame({'TeamId': [1, 2, 3], 'Sex': ['male', 'female', 'male']})
```

```python
df = pd.merge(df, df_gender, on='TeamId', how='left')
```

```python
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
```

```python
if 'TeamId' in df.columns:
    df['TeamId'] = df['TeamId'].map({'C': 0, 'Q': 1, 'S': 2})
else:
    print("Column 'TeamId' not found in the DataFrame.")
```

```python
numeric_df = df.select_dtypes(include=np.number)
```

```python
# Correlation matrix
plt.figure(figsize=(10,6))
```

```python
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
```
